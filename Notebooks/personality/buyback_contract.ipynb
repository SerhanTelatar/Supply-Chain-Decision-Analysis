```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
personality = pd.read_excel("../../Datasets/Personality Survey Data.xlsx")
```

```python
buyback_contract = personality.iloc[51:95, :]
buyback_contract = buyback_contract.iloc[:,2:]
buyback_contract.columns = personality.iloc[50, 2:]
buyback_contract.reset_index(drop=True, inplace = True)

```

```python
buyback_contract.columns
```

```python
buyback_contract
```

```python
faculty = pd.get_dummies(buyback_contract['Faculty'])
faculty = faculty.replace({True: 1, False: 0})
faculty.columns = ['faculty_' + col for col in faculty.columns]
buyback_contract = pd.concat([buyback_contract, faculty], axis=1)
buyback_contract.drop(["Faculty"], axis=1,inplace = True)

buyback_contract
```

```python
program = pd.get_dummies(buyback_contract['Program'])
program = program.replace({True: 1, False: 0, })
program.columns = ['program_' + col for col in program.columns]
buyback_contract = pd.concat([buyback_contract, program], axis=1)
buyback_contract.drop(["Program"], axis=1,inplace = True)
```

```python

Grad_Undergrad = pd.get_dummies(buyback_contract['Grad/ Undergrad'])
Grad_Undergrad = Grad_Undergrad.replace({True: 1, False: 0})
Grad_Undergrad.columns = ['Grad_Undergrad_' + col for col in Grad_Undergrad.columns]
buyback_contract = pd.concat([buyback_contract, Grad_Undergrad], axis=1)
buyback_contract.drop(["Grad/ Undergrad"], axis=1,inplace = True)

```

```python



buyback_contract["ENS 208"] = buyback_contract["ENS 208"].replace({"YES": 1, np.nan: 0}).astype(int)
buyback_contract["MS 401"] = buyback_contract["MS 401"].replace({"YES": 1, np.nan: 0}).astype(int)
buyback_contract["Other Supply Chain Course?"] = buyback_contract["Other Supply Chain Course?"].replace({"YES": 1, np.nan: 0}).astype(int)
buyback_contract
```

```python
buyback_contract.to_csv('../../Datasets/buyback_contract_subject_background_information.csv', index=False)
```

```python
post_experiment = pd.read_excel("../../Datasets/Personality Survey Data.xlsx", sheet_name="Post-Experiment Surveys")
```

```python
post_experiment
```

```python
buyback_contract_post_experiment = post_experiment.iloc[51:95,2:25]

buyback_contract_post_experiment.columns = post_experiment.iloc[50,2:25]
buyback_contract_post_experiment.reset_index(drop=True, inplace=True)
```

```python
buyback_contract_post_experiment_manufacturer = buyback_contract_post_experiment.iloc[:22,12:26]
buyback_contract_post_experiment_manufacturer
```

```python
def changer(feature_name, dataset_name):
    dataset_name[feature_name] = dataset_name[feature_name].replace(
        {
            "Affected a lot": 3,
            "Affected": 2,
            "Affected a little": 1,
            "Didn't affect": 0,
            np.nan: -1,
            "Etkilemedi":0
            
        }
    ).astype(int)
```

```python
for i in buyback_contract_post_experiment_manufacturer.columns:
    changer(i, buyback_contract_post_experiment_manufacturer)
buyback_contract_post_experiment_manufacturer = pd.concat([buyback_contract_post_experiment_manufacturer, buyback_contract_post_experiment.iloc[0:22,0]], axis =1)
buyback_contract_post_experiment_manufacturer
```

```python
buyback_contract_post_experiment_manufacturer.to_csv('../../Datasets/buyback_contract_post_experiment_manufacturer.csv', index=False)
```

```python
buyback_contract_post_experiment_retailer = buyback_contract_post_experiment.iloc[22:,:12]
buyback_contract_post_experiment_retailer
```

```python
for i in buyback_contract_post_experiment_retailer.columns[1:]:
    changer(i, buyback_contract_post_experiment_retailer)

buyback_contract_post_experiment_retailer.reset_index(drop=True, inplace=True)
buyback_contract_post_experiment_retailer
```

```python
buyback_contract_post_experiment_retailer.to_csv("../../Datasets/buyback_contract_post_experiment_retailer.csv", index=False)
```
