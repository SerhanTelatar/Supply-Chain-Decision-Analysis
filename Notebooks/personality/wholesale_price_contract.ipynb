```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
personality = pd.read_excel("../../Datasets/Personality Survey Data.xlsx")
```

```python
personality_wholesale_price = personality.iloc[:, 2:]
personality_wholesale_price = personality_wholesale_price.iloc[2:46, :]
personality_wholesale_price.columns = personality.iloc[1, 2:]
personality_wholesale_price.reset_index(drop=True, inplace=True)
```

```python
personality_wholesale_price["Faculty"].value_counts()
```

```python
personality_wholesale_price.columns
```

```python
faculty = pd.get_dummies(personality_wholesale_price['Faculty'])
faculty = faculty.replace({True: 1, False: 0})
faculty.columns = ['faculty_' + col for col in faculty.columns]
personality_wholesale_price = pd.concat([personality_wholesale_price, faculty], axis=1)
personality_wholesale_price.drop(["Faculty"], axis=1,inplace = True)
```

```python
personality_wholesale_price.columns
```

```python
program = pd.get_dummies(personality_wholesale_price['Program'])
program = program.replace({True: 1, False: 0, })
program.columns = ['program_' + col for col in program.columns]
personality_wholesale_price = pd.concat([personality_wholesale_price, program], axis=1)
personality_wholesale_price.drop(["Program"], axis=1,inplace = True)
```

```python
Grad_Undergrad = pd.get_dummies(personality_wholesale_price['Grad/ Undergrad'])
Grad_Undergrad = Grad_Undergrad.replace({True: 1, False: 0})
Grad_Undergrad.columns = ['Grad_Undergrad_' + col for col in Grad_Undergrad.columns]
personality_wholesale_price = pd.concat([personality_wholesale_price, Grad_Undergrad], axis=1)
personality_wholesale_price.drop(["Grad/ Undergrad"], axis=1,inplace = True)
```

```python
personality_wholesale_price["ENS 208"] = personality_wholesale_price["ENS 208"].replace({"YES": 1, np.nan: 0}).astype(int)
personality_wholesale_price["MS 401"] = personality_wholesale_price["MS 401"].replace({"YES": 1, np.nan: 0}).astype(int)
personality_wholesale_price["Other Supply Chain Course?"] = personality_wholesale_price["Other Supply Chain Course?"].replace({"YES": 1, np.nan: 0}).astype(int)
personality_wholesale_price
```

```python
personality_wholesale_price.columns
```

```python

personality_wholesale_price.to_csv('../../Datasets/wholesale_price_contract_subject_background_information.csv', index=False)
```

```python
data4 = pd.read_excel("../../Datasets/Personality Survey Data.xlsx", sheet_name="Post-Experiment Surveys")
```

```python
wholesale_price_contract_post_experiment = data4.iloc[:, 2:24]
wholesale_price_contract_post_experiment = wholesale_price_contract_post_experiment.iloc[2:46, :]
wholesale_price_contract_post_experiment.columns = data4.iloc[1, 2:24]
wholesale_price_contract_post_experiment.reset_index(drop=True, inplace=True)
```

```python
wholesale_price_contract_post_experiment_manufacturer = wholesale_price_contract_post_experiment.iloc[:22,11:]
wholesale_price_contract_post_experiment_manufacturer
```

```python
wholesale_price_contract_post_experiment_manufacturer.columns
```

```python
def changer(feature_name, dataset_name):
    dataset_name[feature_name] = dataset_name[feature_name].replace(
        {
            "Affected a lot": 3,
            "Affected": 2,
            "Affected a little": 1,
            "Didn't affect": 0,
            np.nan: -1
            
        }
    ).astype(int)
    
    
```

```python
wholesale_price_contract_post_experiment_manufacturer.columns
```

```python
for i in wholesale_price_contract_post_experiment_manufacturer.columns:
    changer(i, wholesale_price_contract_post_experiment_manufacturer)
wholesale_price_contract_post_experiment_manufacturer["PLAYER NAME"] = wholesale_price_contract_post_experiment.iloc[:, 0]
wholesale_price_contract_post_experiment_manufacturer
```

```python
wholesale_price_contract_post_experiment_manufacturer.to_csv('../../Datasets/wholesale_price_contract_post_experiment_manufacturer.csv', index=False)
```

```python
wholesale_price_contract_post_experiment_retailer = wholesale_price_contract_post_experiment.iloc[22:,:11]
wholesale_price_contract_post_experiment_retailer.head()
```

```python
wholesale_price_contract_post_experiment_retailer.columns
```
